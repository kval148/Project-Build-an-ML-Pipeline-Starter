3. What do you notice in the data? Look around and see what you can find.

> - **Missing Data**
>   - Columns `name`, `host_name`, `last_review`, and `reviews_per_month` have missing values.
> - **Data Type Mismatch**
>   - Column `last_review` is an object data type but appears to hold dates in a YYYY-MM-DD format
> - **Outliers**
>   - Several columns have outliers, particularly of note for columns `price`, `minimum_nights`, `number_of_reviews`, and `calculated_host_listings_count`.
>   - `price` column has 5 listings with a price of 0, which is suspicious.

> _After talking to your stakeholders, you decide to consider from a minimum of **`$10`** to a maximum of **`$350`** per night._

4. Fix some of the little problems we have found in the data with the following code:

```python
# Drop outliers
min_price = 10
max_price = 350
idx = df['price'].between(min_price, max_price)
df = df[idx].copy()
# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```

Note how we did not impute missing values. We will do that in the inference pipeline, so we will be able to handle missing values also in production.
